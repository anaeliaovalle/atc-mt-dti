# Write proteins.txt

```python
import numpy as np

fasta_file = "../../data/covid/NC_045512_2_aa_sequences.fasta"
```

```python
def parse_fasta(file_name):
    lines = [item.strip("\n") for item in open(file_name).readlines()]
    seqs = []
    header = lines[0]
    sequence = ""
    for l in lines[1:]:
        # headers start with ">"
        if ">" in l:
            seqs += [(header,sequence)]
            header = l
            sequence = ""
        else:
            sequence += l
    seqs += [(header,sequence)]
    return seqs
```

```python
parse_fasta(fasta_file)
```

```python
dic = {}
for header, seq in parse_fasta(fasta_file):
    # parse accession from header
    acc = header.split(">")[1].split(" ")[0]
    dic[acc] = seq
dic
```

```python
# original file uses " not '
str(dic).replace("'",'"')
```

```python
# write to file
prot_out_file = open("../../data/covid/proteins.txt",'w')
prot_out_file.write(str(dic).replace("'",'"'))
prot_out_file.close()
```

```python
from chembl_webresource_client.new_client import new_client
molecule = new_client.molecule
chembl_ids = []
for d in list(drug_names):
    s = d.split("::")
    #print(s)
    chembl = ""
    for s_ in s:
        if "CHEMBL" in s_:
            chembl = s_
            
    # if there is no CHEMBL ID in the name, we look it up
    if len(chembl) == 0:
        for s_ in s:
            mol = molecule.search(s_)
            if mol:
                chembl = mol[0]['molecule_chembl_id']
                #print(chembl)
                #print(mol)
                break
    print(chembl)
    chembl_ids += [chembl]
```

```python
# Also get the ATC classification
chembl_atc_pairs = []
for c in chembl_ids:
    mol = molecule.search(c)
    if mol:
        atc = mol[0]['atc_classifications']
        if atc:
            print((c,atc[0]))
            chembl_atc_pairs+=[(c,atc[0])]
    
    
```

```python
print("Number of putative COVID drugs with ATC: "+str(len(chembl_atc_pairs)))
print("Total number of COVID drugs: "+str(len(chembl_ids)))
```
