```python
headers = ["Drug","CHEMBL","ATC"]
out_file = pd.DataFrame(columns=headers)
for i in range(len(atc_mapping)):
    dic = {}
    drug = atc_mapping['Drug'][i]
    atc_code = atc_mapping['Class_ID'][i]
    
    mol = molecule.search(drug)
    if mol:
        mol = mol[0]
        lookup_atc = mol['atc_classifications']
        chembl = mol['molecule_chembl_id']
        dic['Drug'] = drug
        dic['CHEMBL'] = chembl
        dic['ATC'] = atc_code
        #dic['ATC'] = lookup_atc
        ## lookup_atc is slightly different, some have 0 and some have 2 codes... 
        ## better to use the ones we had in the first place, I think.
        
        #print(lookup_atc)
        #if lookup_atc and lookup_atc[0][0:2] != atc_code[0:2]:
        print(str(lookup_atc)+";"+ atc_code)
        #print(dic)
        out_file = out_file.append(dic,ignore_index=True)
        
#out_file.to_csv("ATC_Drug_to_CHEMBL.csv",index=False)
```

# Combining all the ATC_Drug_to_CHEMBL
It kept crashing for no reason but would run again if you try the ID it stopped on... It also took so long, so I split them into different bins.

```python
file_list = ["ATC_Drug_to_CHEMBL_1_to_3231.csv",
            "ATC_Drug_to_CHEMBL_3000_to_3500.csv",
            "ATC_Drug_to_CHEMBL_3501_to_4000.csv",
            "ATC_Drug_to_CHEMBL_4001_to_4500.csv",
            "ATC_Drug_to_CHEMBL_4501_to_5000.csv",
            "ATC_Drug_to_CHEMBL_5001_to_5500.csv",
            "ATC_Drug_to_CHEMBL_5501_to_6000.csv",
            "ATC_Drug_to_CHEMBL_6001_to_end.csv"]

headers = ["Drug","CHEMBL","ATC"]
full_out_file = pd.DataFrame(columns=headers)
pandas_files = []
for f in file_list:
    fi = pd.read_csv(f)
    pandas_files += [fi]
full_out_file = pd.concat(pandas_files, ignore_index = True)
```
