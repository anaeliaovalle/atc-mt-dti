```python
import pandas as pd
file_name = "../../src/predict/atc-covid-predictions.txt"
results = pd.read_csv(file_name)
# sort by affinity value
results = results.sort_values(['y_hat','chemid', 'pid'], ascending=False)
results
```

```python
from chembl_webresource_client.new_client import new_client
molecule = new_client.molecule

drug_preferred_names = []
atc_codes = []

for drug in results['chemid']:
    res = molecule.search(drug)[0]
    drug_preferred_name = res['pref_name']
    atc = res['atc_classifications']
    drug_preferred_names += [drug_preferred_name]
    atc_codes += [",".join(atc)]
results['drug preferred name'] = drug_preferred_names
results['atc code'] = atc_codes
results
```

```python
results.drop_duplicates()
```

```python
results.to_csv("atc-covid-predictions-with-drug-name-and-atc.csv",index=False)
```
